# An√°lisis Exploratorio de Datos - DATATON

Este notebook contiene el an√°lisis exploratorio de los 6 datasets CSV disponibles:

1. **AGUA_RESIDUAL_EFLUENTES_2.csv** - Datos de agua residual y efluentes
2. **AGUA_SUBTERRANEA_2.csv** - Datos de agua subterr√°nea
3. **Monitoreo_Agua_Agua_superficial_1.csv** - Datos de agua superficial
4. **Monitoreos_AGUA_EVALUACION_CAUSALIDAD_0.csv** - Evaluaci√≥n de causalidad en agua
5. **Monitoreos_AGUA_EVALUACION_TEMPRANA_0.csv** - Evaluaci√≥n temprana en agua
6. **SUELO_SEDIMENTO_2.csv** - Datos de suelo y sedimento

El objetivo es conocer a fondo la estructura y caracter√≠sticas de cada dataset antes de implementar modelos de machine learning.

```python
# Importar librer√≠as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Configurar estilo de visualizaci√≥n
plt.style.use('default')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print("üìä Librer√≠as importadas correctamente")
print("üîç Listo para comenzar el an√°lisis exploratorio")
```

```python
# Funci√≥n auxiliar para an√°lisis exploratorio completo
def analisis_exploratorio_completo(df, nombre_dataset):
    """
    Funci√≥n que realiza un an√°lisis exploratorio completo de un DataFrame
    """
    print(f"{'='*60}")
    print(f"üîç AN√ÅLISIS EXPLORATORIO: {nombre_dataset}")
    print(f"{'='*60}")
    
    # 1. Informaci√≥n b√°sica del dataset
    print("\nüìä INFORMACI√ìN B√ÅSICA DEL DATASET")
    print("-" * 40)
    print(f"Forma del dataset: {df.shape}")
    print(f"N√∫mero de filas: {df.shape[0]:,}")
    print(f"N√∫mero de columnas: {df.shape[1]}")
    
    # 2. Informaci√≥n de tipos de datos
    print("\nüìã TIPOS DE DATOS")
    print("-" * 40)
    print(df.dtypes.value_counts())
    
    # 3. Valores nulos
    print("\n‚ùå VALORES NULOS")
    print("-" * 40)
    valores_nulos = df.isnull().sum()
    porcentaje_nulos = (valores_nulos / len(df)) * 100
    nulos_df = pd.DataFrame({
        'Columna': valores_nulos.index,
        'Valores_Nulos': valores_nulos.values,
        'Porcentaje': porcentaje_nulos.values
    }).sort_values('Porcentaje', ascending=False)
    
    if nulos_df['Valores_Nulos'].sum() > 0:
        print(nulos_df[nulos_df['Valores_Nulos'] > 0])
    else:
        print("‚úÖ No hay valores nulos en el dataset")
    
    # 4. Primeras y √∫ltimas filas
    print("\nüëÄ PRIMERAS 3 FILAS")
    print("-" * 40)
    print(df.head(3))
    
    print("\nüëÄ √öLTIMAS 3 FILAS")
    print("-" * 40)
    print(df.tail(3))
    
    # 5. Estad√≠sticas descriptivas para columnas num√©ricas
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) > 0:
        print(f"\nüìà ESTAD√çSTICAS DESCRIPTIVAS (Columnas num√©ricas: {len(numeric_cols)})")
        print("-" * 40)
        print(df[numeric_cols].describe())
    
    # 6. Informaci√≥n sobre columnas categ√≥ricas
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns
    if len(categorical_cols) > 0:
        print(f"\nüè∑Ô∏è COLUMNAS CATEG√ìRICAS (Total: {len(categorical_cols)})")
        print("-" * 40)
        for col in categorical_cols[:10]:  # Mostrar solo las primeras 10
            unique_vals = df[col].nunique()
            print(f"{col}: {unique_vals} valores √∫nicos")
            if unique_vals <= 10:
                print(f"   Valores: {df[col].unique()}")
    
    return df, numeric_cols, categorical_cols

# Definir las rutas de los archivos CSV
csv_files = {
    'AGUA_RESIDUAL_EFLUENTES': '../DATA/AGUA_RESIDUAL_EFLUENTES_2.csv',
    'AGUA_SUBTERRANEA': '../DATA/AGUA_SUBTERRANEA_2.csv',
    'AGUA_SUPERFICIAL': '../DATA/Monitoreo_Agua_Agua_superficial_1.csv',
    'EVALUACION_CAUSALIDAD': '../DATA/Monitoreos_AGUA_EVALUACION_CAUSALIDAD_0.csv',
    'EVALUACION_TEMPRANA': '../DATA/Monitoreos_AGUA_EVALUACION_TEMPRANA_0.csv',
    'SUELO_SEDIMENTO': '../DATA/SUELO_SEDIMENTO_2.csv'
}

print("üéØ Funciones auxiliares definidas correctamente")
print("üìÅ Rutas de archivos CSV configuradas")
```

## 1. üíß AGUA RESIDUAL Y EFLUENTES

An√°lisis del dataset de agua residual y efluentes. Este dataset contiene informaci√≥n sobre el monitoreo de aguas residuales y efluentes.

```python
# Cargar dataset de Agua Residual y Efluentes
try:
    df_agua_residual = pd.read_csv(csv_files['AGUA_RESIDUAL_EFLUENTES'])
    print("‚úÖ Dataset de Agua Residual cargado exitosamente")
    
    # Realizar an√°lisis exploratorio completo
    df_agua_residual, numeric_cols_ar, categorical_cols_ar = analisis_exploratorio_completo(
        df_agua_residual, "AGUA RESIDUAL Y EFLUENTES"
    )
    
except FileNotFoundError:
    print("‚ùå Error: No se pudo encontrar el archivo AGUA_RESIDUAL_EFLUENTES_2.csv")
except Exception as e:
    print(f"‚ùå Error al cargar el archivo: {e}")
```

```python
# Visualizaciones para Agua Residual y Efluentes
if 'df_agua_residual' in locals():
    print("üìä CREANDO VISUALIZACIONES PARA AGUA RESIDUAL Y EFLUENTES")
    print("-" * 60)
    
    # 1. Distribuci√≥n de valores nulos
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    fig.suptitle('Agua Residual y Efluentes - An√°lisis Visual', fontsize=16, fontweight='bold')
    
    # Valores nulos por columna
    valores_nulos = df_agua_residual.isnull().sum()
    if valores_nulos.sum() > 0:
        valores_nulos_top = valores_nulos[valores_nulos > 0].head(10)
        axes[0,0].barh(range(len(valores_nulos_top)), valores_nulos_top.values)
        axes[0,0].set_yticks(range(len(valores_nulos_top)))
        axes[0,0].set_yticklabels(valores_nulos_top.index, fontsize=8)
        axes[0,0].set_title('Top 10 Columnas con Valores Nulos')
        axes[0,0].set_xlabel('Cantidad de Valores Nulos')
    else:
        axes[0,0].text(0.5, 0.5, 'No hay valores nulos', ha='center', va='center', transform=axes[0,0].transAxes)
        axes[0,0].set_title('Valores Nulos')
    
    # Distribuci√≥n de tipos de datos
    tipos_datos = df_agua_residual.dtypes.value_counts()
    axes[0,1].pie(tipos_datos.values, labels=tipos_datos.index, autopct='%1.1f%%')
    axes[0,1].set_title('Distribuci√≥n de Tipos de Datos')
    
    # Correlaci√≥n entre variables num√©ricas (si hay suficientes)
    if len(numeric_cols_ar) > 1:
        # Seleccionar una muestra de columnas num√©ricas para la correlaci√≥n
        numeric_sample = numeric_cols_ar[:10] if len(numeric_cols_ar) > 10 else numeric_cols_ar
        corr_matrix = df_agua_residual[numeric_sample].corr()
        
        im = axes[1,0].imshow(corr_matrix, cmap='coolwarm', aspect='auto', vmin=-1, vmax=1)
        axes[1,0].set_xticks(range(len(numeric_sample)))
        axes[1,0].set_yticks(range(len(numeric_sample)))
        axes[1,0].set_xticklabels(numeric_sample, rotation=45, ha='right', fontsize=8)
        axes[1,0].set_yticklabels(numeric_sample, fontsize=8)
        axes[1,0].set_title('Matriz de Correlaci√≥n (Muestra)')
        plt.colorbar(im, ax=axes[1,0])
    else:
        axes[1,0].text(0.5, 0.5, 'Insuficientes variables\nnum√©ricas para correlaci√≥n', 
                       ha='center', va='center', transform=axes[1,0].transAxes)
        axes[1,0].set_title('Matriz de Correlaci√≥n')
    
    # Distribuci√≥n de una variable num√©rica (la primera disponible)
    if len(numeric_cols_ar) > 0:
        col_ejemplo = numeric_cols_ar[0]
        df_agua_residual[col_ejemplo].hist(bins=30, ax=axes[1,1], alpha=0.7)
        axes[1,1].set_title(f'Distribuci√≥n de {col_ejemplo}')
        axes[1,1].set_xlabel(col_ejemplo)
        axes[1,1].set_ylabel('Frecuencia')
    else:
        axes[1,1].text(0.5, 0.5, 'No hay variables\nnum√©ricas disponibles', 
                       ha='center', va='center', transform=axes[1,1].transAxes)
        axes[1,1].set_title('Distribuci√≥n Num√©rica')
    
    plt.tight_layout()
    plt.show()
    
    # Informaci√≥n adicional espec√≠fica del dataset
    print("\nüîç INFORMACI√ìN ESPEC√çFICA DEL DATASET")
    print("-" * 40)
    print(f"Columnas num√©ricas identificadas: {len(numeric_cols_ar)}")
    print(f"Columnas categ√≥ricas identificadas: {len(categorical_cols_ar)}")
    
    if len(numeric_cols_ar) > 0:
        print(f"\nPrimeras 5 columnas num√©ricas:")
        for i, col in enumerate(numeric_cols_ar[:5]):
            print(f"  {i+1}. {col}")
    
    if len(categorical_cols_ar) > 0:
        print(f"\nPrimeras 5 columnas categ√≥ricas:")
        for i, col in enumerate(categorical_cols_ar[:5]):
            print(f"  {i+1}. {col}")
```

## 7. üìä RESUMEN COMPARATIVO DE TODOS LOS DATASETS

Comparaci√≥n general de las caracter√≠sticas principales de todos los datasets analizados.

```python
# Crear resumen comparativo de todos los datasets
print("üìã RESUMEN COMPARATIVO DE TODOS LOS DATASETS")
print("=" * 60)

# Lista de datasets cargados
datasets_info = []

# Verificar qu√© datasets se cargaron exitosamente
if 'df_agua_residual' in locals():
    datasets_info.append({
        'Dataset': 'Agua Residual y Efluentes',
        'Filas': f"{df_agua_residual.shape[0]:,}",
        'Columnas': df_agua_residual.shape[1],
        'Cols_Numericas': len(numeric_cols_ar),
        'Cols_Categoricas': len(categorical_cols_ar),
        'Valores_Nulos': df_agua_residual.isnull().sum().sum(),
        'Memoria_MB': round(df_agua_residual.memory_usage(deep=True).sum() / 1024 / 1024, 2)
    })

if 'df_agua_subterranea' in locals():
    datasets_info.append({
        'Dataset': 'Agua Subterr√°nea',
        'Filas': f"{df_agua_subterranea.shape[0]:,}",
        'Columnas': df_agua_subterranea.shape[1],
        'Cols_Numericas': len(numeric_cols_as),
        'Cols_Categoricas': len(categorical_cols_as),
        'Valores_Nulos': df_agua_subterranea.isnull().sum().sum(),
        'Memoria_MB': round(df_agua_subterranea.memory_usage(deep=True).sum() / 1024 / 1024, 2)
    })

if 'df_agua_superficial' in locals():
    datasets_info.append({
        'Dataset': 'Agua Superficial',
        'Filas': f"{df_agua_superficial.shape[0]:,}",
        'Columnas': df_agua_superficial.shape[1],
        'Cols_Numericas': len(numeric_cols_asup),
        'Cols_Categoricas': len(categorical_cols_asup),
        'Valores_Nulos': df_agua_superficial.isnull().sum().sum(),
        'Memoria_MB': round(df_agua_superficial.memory_usage(deep=True).sum() / 1024 / 1024, 2)
    })

if 'df_eval_causalidad' in locals():
    datasets_info.append({
        'Dataset': 'Evaluaci√≥n Causalidad',
        'Filas': f"{df_eval_causalidad.shape[0]:,}",
        'Columnas': df_eval_causalidad.shape[1],
        'Cols_Numericas': len(numeric_cols_ec),
        'Cols_Categoricas': len(categorical_cols_ec),
        'Valores_Nulos': df_eval_causalidad.isnull().sum().sum(),
        'Memoria_MB': round(df_eval_causalidad.memory_usage(deep=True).sum() / 1024 / 1024, 2)
    })

if 'df_eval_temprana' in locals():
    datasets_info.append({
        'Dataset': 'Evaluaci√≥n Temprana',
        'Filas': f"{df_eval_temprana.shape[0]:,}",
        'Columnas': df_eval_temprana.shape[1],
        'Cols_Numericas': len(numeric_cols_et),
        'Cols_Categoricas': len(categorical_cols_et),
        'Valores_Nulos': df_eval_temprana.isnull().sum().sum(),
        'Memoria_MB': round(df_eval_temprana.memory_usage(deep=True).sum() / 1024 / 1024, 2)
    })

if 'df_suelo_sedimento' in locals():
    datasets_info.append({
        'Dataset': 'Suelo y Sedimento',
        'Filas': f"{df_suelo_sedimento.shape[0]:,}",
        'Columnas': df_suelo_sedimento.shape[1],
        'Cols_Numericas': len(numeric_cols_ss),
        'Cols_Categoricas': len(categorical_cols_ss),
        'Valores_Nulos': df_suelo_sedimento.isnull().sum().sum(),
        'Memoria_MB': round(df_suelo_sedimento.memory_usage(deep=True).sum() / 1024 / 1024, 2)
    })

# Crear DataFrame resumen
df_resumen = pd.DataFrame(datasets_info)
print("\nüìä TABLA RESUMEN:")
print(df_resumen.to_string(index=False))

# Estad√≠sticas generales
print(f"\nüéØ ESTAD√çSTICAS GENERALES:")
print(f"   üìÅ Total de datasets analizados: {len(datasets_info)}")
if len(datasets_info) > 0:
    total_filas = sum([int(info['Filas'].replace(',', '')) for info in datasets_info])
    total_columnas = sum([info['Columnas'] for info in datasets_info])
    total_memoria = sum([info['Memoria_MB'] for info in datasets_info])
    
    print(f"   üìä Total de filas: {total_filas:,}")
    print(f"   üìã Total de columnas: {total_columnas}")
    print(f"   üíæ Memoria total utilizada: {total_memoria:.2f} MB")
    print(f"   üìà Promedio de filas por dataset: {total_filas/len(datasets_info):,.0f}")
    print(f"   üìä Promedio de columnas por dataset: {total_columnas/len(datasets_info):.1f}")
```

```python
# Crear visualizaci√≥n comparativa final
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Comparaci√≥n Visual de Todos los Datasets', fontsize=16, fontweight='bold')

# 1. Comparaci√≥n de n√∫mero de filas
filas_data = [int(info['Filas'].replace(',', '')) for info in datasets_info]
datasets_nombres = [info['Dataset'] for info in datasets_info]

axes[0,0].bar(range(len(datasets_nombres)), filas_data, color='skyblue', alpha=0.7)
axes[0,0].set_xticks(range(len(datasets_nombres)))
axes[0,0].set_xticklabels(datasets_nombres, rotation=45, ha='right', fontsize=10)
axes[0,0].set_title('N√∫mero de Filas por Dataset')
axes[0,0].set_ylabel('Cantidad de Filas')
for i, v in enumerate(filas_data):
    axes[0,0].text(i, v + max(filas_data)*0.01, f'{v:,}', ha='center', va='bottom', fontsize=8)

# 2. Comparaci√≥n de columnas num√©ricas vs categ√≥ricas
numeric_cols = [info['Cols_Numericas'] for info in datasets_info]
categorical_cols = [info['Cols_Categoricas'] for info in datasets_info]
x_pos = np.arange(len(datasets_nombres))

axes[0,1].bar(x_pos - 0.2, numeric_cols, 0.4, label='Num√©ricas', color='lightcoral', alpha=0.7)
axes[0,1].bar(x_pos + 0.2, categorical_cols, 0.4, label='Categ√≥ricas', color='lightgreen', alpha=0.7)
axes[0,1].set_xticks(x_pos)
axes[0,1].set_xticklabels(datasets_nombres, rotation=45, ha='right', fontsize=10)
axes[0,1].set_title('Columnas Num√©ricas vs Categ√≥ricas')
axes[0,1].set_ylabel('Cantidad de Columnas')
axes[0,1].legend()

# 3. Distribuci√≥n de valores nulos
valores_nulos = [info['Valores_Nulos'] for info in datasets_info]
axes[1,0].bar(range(len(datasets_nombres)), valores_nulos, color='salmon', alpha=0.7)
axes[1,0].set_xticks(range(len(datasets_nombres)))
axes[1,0].set_xticklabels(datasets_nombres, rotation=45, ha='right', fontsize=10)
axes[1,0].set_title('Valores Nulos por Dataset')
axes[1,0].set_ylabel('Cantidad de Valores Nulos')

# 4. Uso de memoria
memoria_mb = [info['Memoria_MB'] for info in datasets_info]
axes[1,1].bar(range(len(datasets_nombres)), memoria_mb, color='gold', alpha=0.7)
axes[1,1].set_xticks(range(len(datasets_nombres)))
axes[1,1].set_xticklabels(datasets_nombres, rotation=45, ha='right', fontsize=10)
axes[1,1].set_title('Uso de Memoria por Dataset (MB)')
axes[1,1].set_ylabel('Memoria (MB)')

plt.tight_layout()
plt.show()

print("\n‚úÖ AN√ÅLISIS EXPLORATORIO COMPLETADO")
print("=" * 50)
print("üéØ CONCLUSIONES PRINCIPALES:")
print(f"   ‚Ä¢ Se analizaron {len(datasets_info)} datasets con un total de {sum([int(info['Filas'].replace(',', '')) for info in datasets_info]):,} registros")
print(f"   ‚Ä¢ El dataset m√°s grande es '{datasets_nombres[np.argmax(filas_data)]}' con {max(filas_data):,} filas")
print(f"   ‚Ä¢ El dataset con m√°s valores nulos es '{datasets_nombres[np.argmax(valores_nulos)]}' con {max(valores_nulos):,} valores nulos")
print(f"   ‚Ä¢ Uso total de memoria: {sum(memoria_mb):.2f} MB")
print("\nüöÄ Los datos est√°n listos para el siguiente paso: desarrollo de modelos de machine learning")
```

1. **üèûÔ∏è Agua Superficial** - Calidad de cuerpos de agua
2. **üíß Efluentes (Agua Residual)** - Descargas contaminantes
3. **üå± Suelo y Sedimento** - Contaminaci√≥n del sustrato

### **Objetivo:**
Crear un ranking de "hotspots" por cuenca/distrito/sitio que permita priorizar d√≥nde fiscalizar primero, calculando un **√çndice de Riesgo de Fiscalizaci√≥n (IRF)** basado en se√±ales detectadas en los datos.

```python
# AN√ÅLISIS DETALLADO DE LOS 3 DATASETS CLAVE PARA RRH
print("üö® RADAR DE RIESGO H√çDRICO - AN√ÅLISIS DETALLADO")
print("=" * 60)

# Seleccionar los 3 datasets clave
datasets_rrh = {
    'agua_superficial': df_agua_superficial,
    'efluentes': df_agua_residual,  # Agua residual = efluentes
    'sedimentos': df_suelo_sedimento
}

print("\nüìä DATASETS SELECCIONADOS PARA RRH:")
for nombre, df in datasets_rrh.items():
    print(f"   ‚Ä¢ {nombre.upper()}: {df.shape[0]:,} filas, {df.shape[1]} columnas")

# An√°lisis de columnas disponibles para cada dataset
print("\nüîç AN√ÅLISIS DE COLUMNAS DISPONIBLES:")
print("-" * 40)

for nombre, df in datasets_rrh.items():
    print(f"\nüìã {nombre.upper()}:")
    print(f"   Columnas ({len(df.columns)}): {list(df.columns[:10])}{'...' if len(df.columns) > 10 else ''}")
    
    # Identificar columnas geogr√°ficas
    geo_cols = [col for col in df.columns if any(keyword in col.upper() for keyword in 
                ['COORD', 'LAT', 'LON', 'ESTE', 'NORTE', 'ZONA', 'DEPARTAMENTO', 'PROVINCIA', 'DISTRITO'])]
    if geo_cols:
        print(f"   üó∫Ô∏è  Columnas geogr√°ficas: {geo_cols}")
    
    # Identificar columnas temporales
    temp_cols = [col for col in df.columns if any(keyword in col.upper() for keyword in 
                ['FECHA', 'ANHO', 'A√ëO', 'MES', 'DIA'])]
    if temp_cols:
        print(f"   üìÖ Columnas temporales: {temp_cols}")
    
    # Identificar columnas de contaminantes/par√°metros
    param_cols = [col for col in df.columns if any(keyword in col.upper() for keyword in 
                 ['PH', 'OXIGENO', 'DBO', 'DQO', 'METAL', 'PLOMO', 'MERCURIO', 'ARSENICO', 'COLIFORM', 'NITRATO'])]
    if param_cols:
        print(f"   üß™ Columnas de par√°metros: {param_cols}")
    
    # Contar valores nulos
    nulos_pct = (df.isnull().sum().sum() / (df.shape[0] * df.shape[1])) * 100
    print(f"   ‚ùå Porcentaje de valores nulos: {nulos_pct:.2f}%")
```

## üó∫Ô∏è AN√ÅLISIS GEOESPACIAL Y TEMPORAL

An√°lisis de la distribuci√≥n geogr√°fica y temporal de los datos para identificar patrones espaciales y estacionales.

```python
# AN√ÅLISIS GEOESPACIAL Y TEMPORAL
print("üó∫Ô∏è AN√ÅLISIS GEOESPACIAL Y TEMPORAL")
print("=" * 50)

# Funci√≥n para analizar distribuci√≥n geogr√°fica y temporal
def analizar_distribucion_espaciotemporal(df, nombre_dataset):
    print(f"\nüìç {nombre_dataset.upper()}")
    print("-" * 30)
    
    # An√°lisis temporal
    if 'ANHO' in df.columns:
        print("üìÖ Distribuci√≥n temporal:")
        anhos = df['ANHO'].value_counts().sort_index()
        print(f"   ‚Ä¢ Rango de a√±os: {anhos.index.min()} - {anhos.index.max()}")
        print(f"   ‚Ä¢ A√±os con m√°s registros: {anhos.head(3).index.tolist()}")
        
        # An√°lisis mensual
        if 'MES' in df.columns:
            meses = df['MES'].value_counts().sort_index()
            print(f"   ‚Ä¢ Meses con m√°s registros: {meses.head(3).index.tolist()}")
    
    # An√°lisis geogr√°fico por coordinaci√≥n
    if 'COORDINACION' in df.columns:
        print("\nüèõÔ∏è Distribuci√≥n por Coordinaci√≥n:")
        coord_counts = df['COORDINACION'].value_counts()
        print(f"   ‚Ä¢ Total de coordinaciones: {len(coord_counts)}")
        print(f"   ‚Ä¢ Top 5 coordinaciones:")
        for i, (coord, count) in enumerate(coord_counts.head(5).items(), 1):
            print(f"      {i}. {coord}: {count:,} registros")
    
    # An√°lisis geogr√°fico por zona
    if 'TXZONA' in df.columns:
        print("\nüåç Distribuci√≥n por Zona:")
        zona_counts = df['TXZONA'].value_counts()
        print(f"   ‚Ä¢ Total de zonas: {len(zona_counts)}")
        print(f"   ‚Ä¢ Top 5 zonas:")
        for i, (zona, count) in enumerate(zona_counts.head(5).items(), 1):
            print(f"      {i}. {zona}: {count:,} registros")
    
    # An√°lisis de coordenadas geogr√°ficas
    if 'COORD_ESTE' in df.columns and 'COORD_NORTE' in df.columns:
        coords_validas = df[['COORD_ESTE', 'COORD_NORTE']].dropna()
        print(f"\nüìê Coordenadas geogr√°ficas:")
        print(f"   ‚Ä¢ Registros con coordenadas v√°lidas: {len(coords_validas):,} ({len(coords_validas)/len(df)*100:.1f}%)")
        if len(coords_validas) > 0:
            print(f"   ‚Ä¢ Rango Este: {coords_validas['COORD_ESTE'].min():.0f} - {coords_validas['COORD_ESTE'].max():.0f}")
            print(f"   ‚Ä¢ Rango Norte: {coords_validas['COORD_NORTE'].min():.0f} - {coords_validas['COORD_NORTE'].max():.0f}")
    
    return df

# Aplicar an√°lisis a cada dataset
for nombre, df in datasets_rrh.items():
    analizar_distribucion_espaciotemporal(df, nombre)
```

## üéØ VARIABLES CR√çTICAS PARA EL √çNDICE DE RIESGO DE FISCALIZACI√ìN (IRF)

Identificaci√≥n de las variables m√°s relevantes para calcular el riesgo de fiscalizaci√≥n en cada dataset.

```python
# IDENTIFICACI√ìN DE VARIABLES CR√çTICAS PARA IRF
print("üéØ VARIABLES CR√çTICAS PARA EL √çNDICE DE RIESGO DE FISCALIZACI√ìN (IRF)")
print("=" * 70)

# Definir categor√≠as de variables para el IRF
variables_irf = {
    'GEOGR√ÅFICAS': ['COORDINACION', 'TXZONA', 'COORD_ESTE', 'COORD_NORTE', 'TXUBIGEO'],
    'TEMPORALES': ['ANHO', 'MES', 'FECHAINI', 'FECHAFIN', 'FECHA_PTO'],
    'ADMINISTRATIVAS': ['EXPEDIENTE', 'CUC', 'TXORIGEN'],
    'CONTAMINANTES': [],  # Se llenar√° din√°micamente
    'MEDICIONES': []      # Se llenar√° din√°micamente
}

# Funci√≥n para identificar variables cr√≠ticas en cada dataset
def identificar_variables_criticas(df, nombre_dataset):
    print(f"\nüîç {nombre_dataset.upper()}")
    print("-" * 40)
    
    # Identificar todas las columnas disponibles
    columnas_disponibles = df.columns.tolist()
    
    # Identificar variables de contaminantes y mediciones
    contaminantes = []
    mediciones = []
    
    for col in columnas_disponibles:
        col_upper = col.upper()
        
        # Contaminantes espec√≠ficos
        if any(contaminante in col_upper for contaminante in 
               ['METAL', 'PLOMO', 'MERCURIO', 'ARSENICO', 'CADMIO', 'CROMO', 'ZINC', 
                'COBRE', 'HIERRO', 'MANGANESO', 'NIQUEL']):
            contaminantes.append(col)
        
        # Par√°metros de medici√≥n
        elif any(param in col_upper for param in 
                ['PH', 'OXIGENO', 'DBO', 'DQO', 'COLIFORM', 'NITRATO', 'FOSFATO',
                 'TURBIDEZ', 'CONDUCTIVIDAD', 'TEMPERATURA', 'SOLIDOS']):
            mediciones.append(col)
        
        # Otros par√°metros espec√≠ficos por tipo
        elif 'SEDIMENTO' in nombre_dataset.upper() and any(param in col_upper for param in 
                ['SHAKE', 'FLASK', 'SOLUBILITY']):
            mediciones.append(col)
    
    # Mostrar variables disponibles por categor√≠a
    print("üìä Variables disponibles por categor√≠a:")
    
    for categoria, vars_base in variables_irf.items():
        vars_encontradas = [col for col in vars_base if col in columnas_disponibles]
        
        if categoria == 'CONTAMINANTES':
            vars_encontradas = contaminantes
        elif categoria == 'MEDICIONES':
            vars_encontradas = mediciones
        
        if vars_encontradas:
            print(f"   ‚Ä¢ {categoria}: {len(vars_encontradas)} variables")
            for var in vars_encontradas[:5]:  # Mostrar solo las primeras 5
                print(f"     - {var}")
            if len(vars_encontradas) > 5:
                print(f"     ... y {len(vars_encontradas) - 5} m√°s")
    
    # Calcular completitud de datos para variables cr√≠ticas
    print(f"\nüìà Completitud de datos (% de valores no nulos):")
    variables_criticas = []
    
    # Agregar variables geogr√°ficas cr√≠ticas
    for var in ['COORDINACION', 'COORD_ESTE', 'COORD_NORTE']:
        if var in columnas_disponibles:
            completitud = (1 - df[var].isnull().sum() / len(df)) * 100
            print(f"   ‚Ä¢ {var}: {completitud:.1f}%")
            if completitud > 50:  # Solo considerar variables con >50% de completitud
                variables_criticas.append(var)
    
    # Agregar variables de contaminantes
    for var in contaminantes[:3]:  # Solo las primeras 3
        if var in df.columns:
            completitud = (1 - df[var].isnull().sum() / len(df)) * 100
            print(f"   ‚Ä¢ {var}: {completitud:.1f}%")
            if completitud > 20:  # Umbral m√°s bajo para contaminantes
                variables_criticas.append(var)
    
    return variables_criticas, contaminantes, mediciones

# Analizar cada dataset
resultados_irf = {}
for nombre, df in datasets_rrh.items():
    variables_criticas, contaminantes, mediciones = identificar_variables_criticas(df, nombre)
    resultados_irf[nombre] = {
        'variables_criticas': variables_criticas,
        'contaminantes': contaminantes,
        'mediciones': mediciones,
        'total_registros': len(df)
    }

# Resumen de variables cr√≠ticas identificadas
print(f"\nüìã RESUMEN DE VARIABLES CR√çTICAS PARA IRF:")
print("-" * 50)
for nombre, resultado in resultados_irf.items():
    print(f"\n{nombre.upper()}:")
    print(f"   ‚Ä¢ Variables cr√≠ticas identificadas: {len(resultado['variables_criticas'])}")
    print(f"   ‚Ä¢ Contaminantes detectados: {len(resultado['contaminantes'])}")
    print(f"   ‚Ä¢ Par√°metros de medici√≥n: {len(resultado['mediciones'])}")
    print(f"   ‚Ä¢ Total de registros: {resultado['total_registros']:,}")
```

## üß™ AN√ÅLISIS DE CALIDAD Y CONTAMINANTES

An√°lisis detallado de los niveles de contaminaci√≥n y calidad del agua para identificar zonas de alto riesgo.

```python
# AN√ÅLISIS DETALLADO DE CONTAMINANTES Y CALIDAD
print("üß™ AN√ÅLISIS DE CALIDAD Y CONTAMINANTES")
print("=" * 50)

# Funci√≥n para analizar contaminantes por dataset
def analizar_contaminantes_dataset(df, nombre_dataset, contaminantes_cols):
    print(f"\nüî¨ {nombre_dataset.upper()}")
    print("-" * 40)
    
    if not contaminantes_cols:
        print("   ‚ùå No se detectaron columnas espec√≠ficas de contaminantes")
        
        # Buscar columnas que podr√≠an contener informaci√≥n de contaminantes
        cols_posibles = [col for col in df.columns if 
                        any(palabra in col.upper() for palabra in 
                           ['METAL', 'TOTAL', 'DISUELTO', 'BTEX', 'OTROS', 'RS'])]
        
        if cols_posibles:
            print(f"   üîç Columnas potenciales de contaminantes: {cols_posibles}")
            contaminantes_cols = cols_posibles[:3]  # Tomar las primeras 3
    
    # Analizar estad√≠sticas de contaminantes
    if contaminantes_cols:
        print(f"   üìä An√°lisis de {len(contaminantes_cols)} variables de contaminantes:")
        
        for col in contaminantes_cols:
            if col in df.columns:
                # Convertir a num√©rico si es posible
                try:
                    datos_numericos = pd.to_numeric(df[col], errors='coerce')
                    datos_validos = datos_numericos.dropna()
                    
                    if len(datos_validos) > 0:
                        print(f"\n     üìà {col}:")
                        print(f"        ‚Ä¢ Registros v√°lidos: {len(datos_validos):,} ({len(datos_validos)/len(df)*100:.1f}%)")
                        print(f"        ‚Ä¢ Rango: {datos_validos.min():.2f} - {datos_validos.max():.2f}")
                        print(f"        ‚Ä¢ Media: {datos_validos.mean():.2f}")
                        print(f"        ‚Ä¢ Mediana: {datos_validos.median():.2f}")
                        
                        # Identificar valores extremos (potenciales contaminaciones altas)
                        q95 = datos_validos.quantile(0.95)
                        valores_altos = datos_validos[datos_validos > q95]
                        if len(valores_altos) > 0:
                            print(f"        ‚Ä¢ Valores en percentil 95+: {len(valores_altos)} registros (>{q95:.2f})")
                    else:
                        print(f"\n     ‚ùå {col}: No hay datos num√©ricos v√°lidos")
                        
                except Exception as e:
                    print(f"\n     ‚ö†Ô∏è {col}: Error al procesar datos - {str(e)[:50]}")

# Aplicar an√°lisis de contaminantes a cada dataset
for nombre, df in datasets_rrh.items():
    contaminantes = resultados_irf[nombre]['contaminantes']
    analizar_contaminantes_dataset(df, nombre, contaminantes)

# An√°lisis de patrones geogr√°ficos de contaminaci√≥n
print(f"\nüó∫Ô∏è AN√ÅLISIS GEOGR√ÅFICO DE CONTAMINACI√ìN")
print("-" * 50)

for nombre, df in datasets_rrh.items():
    if 'COORDINACION' in df.columns:
        print(f"\nüìç {nombre.upper()} - Top 5 coordinaciones con m√°s registros:")
        coord_counts = df['COORDINACION'].value_counts().head(5)
        for i, (coord, count) in enumerate(coord_counts.items(), 1):
            porcentaje = (count / len(df)) * 100
            print(f"   {i}. {coord}: {count:,} registros ({porcentaje:.1f}%)")

# An√°lisis temporal de tendencias
print(f"\nüìÖ AN√ÅLISIS TEMPORAL DE TENDENCIAS")
print("-" * 50)

for nombre, df in datasets_rrh.items():
    if 'ANHO' in df.columns:
        print(f"\nüìä {nombre.upper()} - Evoluci√≥n temporal:")
        anhos_counts = df['ANHO'].value_counts().sort_index()
        anhos_recientes = anhos_counts.tail(5)
        
        print(f"   ‚Ä¢ A√±os m√°s recientes:")
        for anho, count in anhos_recientes.items():
            porcentaje = (count / len(df)) * 100
            print(f"     - {anho}: {count:,} registros ({porcentaje:.1f}%)")
        
        # Calcular tendencia
        if len(anhos_recientes) > 2:
            tendencia = anhos_recientes.iloc[-1] - anhos_recientes.iloc[0]
            if tendencia > 0:
                print(f"   üìà Tendencia: CRECIENTE (+{tendencia:,} registros)")
            else:
                print(f"   üìâ Tendencia: DECRECIENTE ({tendencia:,} registros)")
```

## üéØ IDENTIFICACI√ìN DE HOTSPOTS POR REGI√ìN

Evaluaci√≥n de completitud de datos y densidad de registros por regi√≥n para identificar zonas prioritarias para fiscalizaci√≥n.

```python
# IDENTIFICACI√ìN DE HOTSPOTS POR REGI√ìN
print("üéØ IDENTIFICACI√ìN DE HOTSPOTS POR REGI√ìN")
print("=" * 50)

# Funci√≥n para calcular el √≠ndice de riesgo por regi√≥n
def calcular_hotspots_region(df, nombre_dataset):
    print(f"\nüó∫Ô∏è {nombre_dataset.upper()}")
    print("-" * 40)
    
    if 'COORDINACION' not in df.columns:
        print("   ‚ùå No hay informaci√≥n de coordinaci√≥n disponible")
        return None
    
    # Agrupar por coordinaci√≥n
    hotspots = df.groupby('COORDINACION').agg({
        'EXPEDIENTE': 'count',  # N√∫mero total de registros
        'ANHO': ['min', 'max'],  # Rango temporal
        'COORD_ESTE': 'count',   # Registros con coordenadas
        'COORD_NORTE': 'count'
    }).round(2)
    
    # Simplificar nombres de columnas
    hotspots.columns = ['Total_Registros', 'A√±o_Min', 'A√±o_Max', 'Coord_Este_Count', 'Coord_Norte_Count']
    
    # Calcular m√©tricas de riesgo
    hotspots['A√±os_Monitoreados'] = hotspots['A√±o_Max'] - hotspots['A√±o_Min'] + 1
    hotspots['Registros_por_A√±o'] = hotspots['Total_Registros'] / hotspots['A√±os_Monitoreados']
    hotspots['Pct_Con_Coordenadas'] = (hotspots['Coord_Este_Count'] / hotspots['Total_Registros']) * 100
    
    # Calcular puntuaci√≥n de riesgo (mayor = m√°s riesgo)
    # Normalizar m√©tricas (0-1)
    hotspots['Score_Frecuencia'] = (hotspots['Total_Registros'] - hotspots['Total_Registros'].min()) / (hotspots['Total_Registros'].max() - hotspots['Total_Registros'].min())
    hotspots['Score_Intensidad'] = (hotspots['Registros_por_A√±o'] - hotspots['Registros_por_A√±o'].min()) / (hotspots['Registros_por_A√±o'].max() - hotspots['Registros_por_A√±o'].min())
    hotspots['Score_Geolocalizaci√≥n'] = hotspots['Pct_Con_Coordenadas'] / 100
    
    # √çndice de Riesgo de Fiscalizaci√≥n (IRF)
    hotspots['IRF'] = (
        hotspots['Score_Frecuencia'] * 0.4 +      # 40% frecuencia de registros
        hotspots['Score_Intensidad'] * 0.4 +      # 40% intensidad de monitoreo
        hotspots['Score_Geolocalizaci√≥n'] * 0.2   # 20% calidad de geolocalizaci√≥n
    ) * 100  # Escalar a 0-100
    
    # Clasificar riesgo
    hotspots['Categoria_Riesgo'] = pd.cut(hotspots['IRF'], 
                                         bins=[0, 25, 50, 75, 100], 
                                         labels=['BAJO', 'MEDIO', 'ALTO', 'CR√çTICO'],
                                         include_lowest=True)
    
    # Ordenar por IRF descendente
    hotspots_ordenados = hotspots.sort_values('IRF', ascending=False)
    
    print(f"   üìä Resumen de hotspots identificados:")
    print(f"      ‚Ä¢ Total de coordinaciones: {len(hotspots_ordenados)}")
    
    # Distribuci√≥n por categor√≠a de riesgo
    risk_distribution = hotspots_ordenados['Categoria_Riesgo'].value_counts()
    for categoria, count in risk_distribution.items():
        pct = (count / len(hotspots_ordenados)) * 100
        print(f"      ‚Ä¢ {categoria}: {count} coordinaciones ({pct:.1f}%)")
    
    print(f"\n   üö® TOP 10 HOTSPOTS (Mayor IRF):")
    top_hotspots = hotspots_ordenados.head(10)
    
    for i, (coordinacion, datos) in enumerate(top_hotspots.iterrows(), 1):
        print(f"      {i:2d}. {coordinacion}")
        print(f"          IRF: {datos['IRF']:.1f} | Categor√≠a: {datos['Categoria_Riesgo']}")
        print(f"          Registros: {datos['Total_Registros']:,} | A√±os: {datos['A√±os_Monitoreados']}")
        print(f"          Intensidad: {datos['Registros_por_A√±o']:.1f} reg/a√±o")
    
    return hotspots_ordenados

# Calcular hotspots para cada dataset
hotspots_resultados = {}
for nombre, df in datasets_rrh.items():
    hotspots = calcular_hotspots_region(df, nombre)
    if hotspots is not None:
        hotspots_resultados[nombre] = hotspots

# An√°lisis consolidado de hotspots
print(f"\nüéØ AN√ÅLISIS CONSOLIDADO DE HOTSPOTS")
print("=" * 50)

if hotspots_resultados:
    # Encontrar coordinaciones que aparecen en m√∫ltiples datasets
    todas_coordinaciones = set()
    for hotspots in hotspots_resultados.values():
        todas_coordinaciones.update(hotspots.index)
    
    coordinaciones_criticas = []
    for coord in todas_coordinaciones:
        datasets_presentes = []
        irf_promedio = []
        
        for nombre, hotspots in hotspots_resultados.items():
            if coord in hotspots.index:
                datasets_presentes.append(nombre)
                irf_promedio.append(hotspots.loc[coord, 'IRF'])
        
        if len(datasets_presentes) >= 2:  # Coordinaci√≥n presente en al menos 2 datasets
            coordinaciones_criticas.append({
                'coordinacion': coord,
                'datasets': datasets_presentes,
                'irf_promedio': np.mean(irf_promedio),
                'num_datasets': len(datasets_presentes)
            })
    
    # Ordenar coordinaciones cr√≠ticas por IRF promedio
    coordinaciones_criticas.sort(key=lambda x: x['irf_promedio'], reverse=True)
    
    print(f"\nüö® COORDINACIONES CR√çTICAS (presentes en m√∫ltiples datasets):")
    print(f"    Total identificadas: {len(coordinaciones_criticas)}")
    
    for i, coord_info in enumerate(coordinaciones_criticas[:10], 1):
        print(f"\n    {i:2d}. {coord_info['coordinacion']}")
        print(f"        IRF Promedio: {coord_info['irf_promedio']:.1f}")
        print(f"        Presente en: {coord_info['num_datasets']} datasets")
        print(f"        Datasets: {', '.join(coord_info['datasets'])}")

print(f"\n‚úÖ HOTSPOTS IDENTIFICADOS PARA PRIORIZACI√ìN DE FISCALIZACI√ìN")
```

```python
# üîç AN√ÅLISIS DETALLADO DE ESTRUCTURA DE COLUMNAS
print("üîç AN√ÅLISIS DETALLADO DE ESTRUCTURA DE COLUMNAS")
print("=" * 60)

def analizar_columnas_dataset(df, nombre):
    """Analiza las columnas de un dataset y las categoriza"""
    
    print(f"\nüìã {nombre} ({df.shape[0]:,} filas, {df.shape[1]} columnas)")
    print("-" * 50)
    
    columnas = df.columns.tolist()
    
    # Categorizar columnas
    categorias = {
        'IDENTIFICACI√ìN': [],
        'GEOGR√ÅFICAS': [],
        'TEMPORALES': [],
        'ADMINISTRATIVAS': [],
        'RESULTADOS/MEDICIONES': [],
        'OTRAS': []
    }
    
    for col in columnas:
        col_upper = col.upper()
        
        # Identificaci√≥n
        if any(palabra in col_upper for palabra in ['ID', 'CODIGO', 'CUC', 'EXPEDIENTE']):
            categorias['IDENTIFICACI√ìN'].append(col)
        
        # Geogr√°ficas
        elif any(palabra in col_upper for palabra in ['COORD', 'LAT', 'LON', 'ESTE', 'NORTE', 'ZONA', 'DEPARTAMENTO', 'PROVINCIA', 'DISTRITO', 'UBIGEO', 'DATUM']):
            categorias['GEOGR√ÅFICAS'].append(col)
        
        # Temporales
        elif any(palabra in col_upper for palabra in ['FECHA', 'ANHO', 'A√ëO', 'MES', 'DIA']):
            categorias['TEMPORALES'].append(col)
        
        # Administrativas
        elif any(palabra in col_upper for palabra in ['COORDINACION', 'ORIGEN', 'OBSERVACIONES', 'EVALUACION', 'CUMPLE']):
            categorias['ADMINISTRATIVAS'].append(col)
        
        # Resultados/Mediciones (par√°metros qu√≠micos, f√≠sicos, biol√≥gicos)
        elif any(palabra in col_upper for palabra in ['RESULTADO', 'VALOR', 'UNIDAD', 'LIMITE', 'ECA', 'LMP', 'PARAMETRO']):
            categorias['RESULTADOS/MEDICIONES'].append(col)
        
        else:
            categorias['OTRAS'].append(col)
    
    # Mostrar categorizaci√≥n
    for categoria, cols in categorias.items():
        if cols:
            print(f"\n   üìä {categoria} ({len(cols)} columnas):")
            for col in cols[:5]:  # Mostrar solo las primeras 5
                non_null = df[col].notna().sum()
                completitud = (non_null / len(df)) * 100
                print(f"      ‚Ä¢ {col}: {completitud:.1f}% completo")
            if len(cols) > 5:
                print(f"      ... y {len(cols) - 5} m√°s")
    
    return categorias

# Analizar cada dataset
datasets_columnas = {}
for nombre, df in datasets_oefa.items():
    categorias = analizar_columnas_dataset(df, nombre)
    datasets_columnas[nombre] = categorias
```

```python
# üîó AN√ÅLISIS DE COLUMNAS COMUNES Y ESPEC√çFICAS
print("üîó AN√ÅLISIS DE COLUMNAS COMUNES Y ESPEC√çFICAS")
print("=" * 60)

# Obtener todas las columnas de todos los datasets
todas_columnas = set()
columnas_por_dataset = {}

for nombre, df in datasets_oefa.items():
    columnas_por_dataset[nombre] = set(df.columns)
    todas_columnas.update(df.columns)

print(f"üìä Total de columnas √∫nicas entre todos los datasets: {len(todas_columnas)}")

# Encontrar columnas comunes a todos los datasets
columnas_comunes = set(columnas_por_dataset[list(columnas_por_dataset.keys())[0]])
for columnas in columnas_por_dataset.values():
    columnas_comunes = columnas_comunes.intersection(columnas)

print(f"\n‚úÖ COLUMNAS COMUNES A TODOS LOS DATASETS ({len(columnas_comunes)}):")
for col in sorted(columnas_comunes):
    print(f"   ‚Ä¢ {col}")

# Encontrar columnas espec√≠ficas por dataset
print(f"\nüîç COLUMNAS ESPEC√çFICAS POR DATASET:")
for nombre, columnas in columnas_por_dataset.items():
    columnas_especificas = columnas - columnas_comunes
    if columnas_especificas:
        print(f"\n   üìã {nombre} - {len(columnas_especificas)} columnas espec√≠ficas:")
        for col in sorted(list(columnas_especificas)[:10]):  # Mostrar solo las primeras 10
            print(f"      ‚Ä¢ {col}")
        if len(columnas_especificas) > 10:
            print(f"      ... y {len(columnas_especificas) - 10} m√°s")

# An√°lisis de completitud de columnas comunes
print(f"\nüìà COMPLETITUD DE COLUMNAS COMUNES:")
completitud_comunes = {}
for col in sorted(columnas_comunes):
    completitud_comunes[col] = {}
    for nombre, df in datasets_oefa.items():
        non_null = df[col].notna().sum()
        completitud = (non_null / len(df)) * 100
        completitud_comunes[col][nombre] = completitud
    
    # Calcular promedio de completitud
    promedio_completitud = np.mean(list(completitud_comunes[col].values()))
    print(f"   ‚Ä¢ {col}: {promedio_completitud:.1f}% promedio")
    
    # Mostrar detalles si hay diferencias significativas
    valores = list(completitud_comunes[col].values())
    if max(valores) - min(valores) > 20:  # Diferencia >20%
        print(f"     ‚îî‚îÄ Variaci√≥n: {min(valores):.1f}% - {max(valores):.1f}%")
```

```python
# üß™ IDENTIFICACI√ìN DE COLUMNAS DE RESULTADOS DE MUESTRAS
print("üß™ IDENTIFICACI√ìN DE COLUMNAS DE RESULTADOS DE MUESTRAS")
print("=" * 60)

def identificar_columnas_resultados(df, nombre_dataset):
    """Identifica las columnas que contienen resultados de an√°lisis"""
    
    print(f"\nüî¨ {nombre_dataset}")
    print("-" * 40)
    
    columnas = df.columns.tolist()
    
    # Categorizar columnas de resultados
    columnas_resultados = {
        'PARAMETROS_FISICOQUIMICOS': [],
        'METALES_PESADOS': [],
        'MICROBIOLOGICOS': [],
        'ORGANICOS': [],
        'LIMITES_NORMATIVOS': [],
        'UNIDADES': [],
        'OTROS_RESULTADOS': []
    }
    
    for col in columnas:
        col_upper = col.upper()
        
        # Par√°metros fisicoqu√≠micos b√°sicos
        if any(param in col_upper for param in ['PH', 'OXIGENO', 'DBO', 'DQO', 'TURBIDEZ', 'CONDUCTIVIDAD', 
                                                'TEMPERATURA', 'SOLIDOS', 'NITRATO', 'FOSFATO', 'SULFATO']):
            columnas_resultados['PARAMETROS_FISICOQUIMICOS'].append(col)
        
        # Metales pesados
        elif any(metal in col_upper for metal in ['PLOMO', 'MERCURIO', 'ARSENICO', 'CADMIO', 'CROMO', 
                                                  'ZINC', 'COBRE', 'HIERRO', 'MANGANESO', 'NIQUEL', 'METAL']):
            columnas_resultados['METALES_PESADOS'].append(col)
        
        # Microbiol√≥gicos
        elif any(micro in col_upper for micro in ['COLIFORM', 'BACTERIA', 'ESCHERICHIA', 'ENTEROCOCO']):
            columnas_resultados['MICROBIOLOGICOS'].append(col)
        
        # Compuestos org√°nicos
        elif any(org in col_upper for org in ['BTEX', 'HIDROCARBUROS', 'PESTICIDAS', 'SOLVENTES']):
            columnas_resultados['ORGANICOS'].append(col)
        
        # L√≠mites normativos
        elif any(limite in col_upper for limite in ['LMP', 'ECA', 'LIMITE', 'ESTANDAR']):
            columnas_resultados['LIMITES_NORMATIVOS'].append(col)
        
        # Unidades de medida
        elif any(unidad in col_upper for unidad in ['UNIDAD', 'UNIT']):
            columnas_resultados['UNIDADES'].append(col)
        
        # Otros resultados (valores num√©ricos que podr√≠an ser mediciones)
        elif ('RESULTADO' in col_upper or 'VALOR' in col_upper or 
              (df[col].dtype in ['float64', 'int64'] and col_upper not in ['ANHO', 'MES', 'DIA'])):
            columnas_resultados['OTROS_RESULTADOS'].append(col)
    
    # Mostrar resultados
    total_resultados = sum(len(cols) for cols in columnas_resultados.values())
    print(f"   üìä Total de columnas de resultados identificadas: {total_resultados}")
    
    for categoria, cols in columnas_resultados.items():
        if cols:
            print(f"\n   üß™ {categoria} ({len(cols)} columnas):")
            for col in cols[:5]:  # Mostrar solo las primeras 5
                non_null = df[col].notna().sum()
                completitud = (non_null / len(df)) * 100
                
                # Mostrar ejemplo de valores para entender el contenido
                ejemplo_valores = df[col].dropna().head(3).tolist()
                print(f"      ‚Ä¢ {col}: {completitud:.1f}% completo")
                print(f"        ‚îî‚îÄ Ejemplos: {ejemplo_valores}")
                
            if len(cols) > 5:
                print(f"      ... y {len(cols) - 5} m√°s")
    
    return columnas_resultados

# Analizar cada dataset para encontrar columnas de resultados
resultados_por_dataset = {}
for nombre, df in datasets_oefa.items():
    resultados = identificar_columnas_resultados(df, nombre)
    resultados_por_dataset[nombre] = resultados
```

```python
# üîß PROCESAMIENTO INDIVIDUAL DE DATASETS OEFA PARA HACKATON
print("üîß PROCESAMIENTO INDIVIDUAL DE DATASETS OEFA PARA HACKATON")
print("=" * 70)

def procesar_dataset_oefa(df, nombre_dataset):
    """
    Procesa un dataset OEFA manteniendo solo columnas importantes
    y filtrando registros con coordenadas v√°lidas
    """
    print(f"\nüìä PROCESANDO: {nombre_dataset}")
    print("-" * 50)
    
    original_filas = len(df)
    print(f"   üìã Registros originales: {original_filas:,}")
    print(f"   üìã Columnas originales: {df.shape[1]}")
    
    # Definir categor√≠as de columnas importantes
    columnas_importantes = {
        'identificacion': [],
        'geograficas': [],
        'temporales': [],
        'resultados': [],
        'administrativas': []
    }
    
    # Categorizar columnas
    for col in df.columns:
        col_upper = col.upper()
        
        # Identificaci√≥n
        if any(palabra in col_upper for palabra in ['EXPEDIENTE', 'CUC', 'ID', 'CODIGO']):
            columnas_importantes['identificacion'].append(col)
        
        # Geogr√°ficas (CR√çTICAS para el hackaton)
        elif any(palabra in col_upper for palabra in ['COORD_ESTE', 'COORD_NORTE', 'ESTE', 'NORTE', 'LAT', 'LON', 'ZONA', 'DATUM', 'COORDINACION', 'UBIGEO']):
            columnas_importantes['geograficas'].append(col)
        
        # Temporales
        elif any(palabra in col_upper for palabra in ['FECHA', 'ANHO', 'A√ëO', 'MES', 'DIA']):
            columnas_importantes['temporales'].append(col)
        
        # Resultados (par√°metros, valores, l√≠mites)
        elif any(palabra in col_upper for palabra in ['PARAMETRO', 'VALOR', 'RESULTADO', 'LIMITE', 'ECA', 'LMP', 'UNIDAD']):
            columnas_importantes['resultados'].append(col)
        
        # Administrativas importantes
        elif any(palabra in col_upper for palabra in ['PUNTO_MUESTREO', 'ORIGEN', 'OBSERVACIONES']):
            columnas_importantes['administrativas'].append(col)
    
    # Mostrar columnas encontradas
    total_importantes = sum(len(cols) for cols in columnas_importantes.values())
    print(f"   üîç Columnas importantes encontradas: {total_importantes}")
    
    for categoria, cols in columnas_importantes.items():
        if cols:
            print(f"     ‚Ä¢ {categoria.upper()}: {len(cols)} columnas")
            for col in cols[:3]:  # Mostrar solo las primeras 3
                completitud = (df[col].notna().sum() / len(df)) * 100
                print(f"       - {col}: {completitud:.1f}%")
            if len(cols) > 3:
                print(f"       ... y {len(cols) - 3} m√°s")
    
    # Crear lista de todas las columnas importantes
    todas_importantes = []
    for cols in columnas_importantes.values():
        todas_importantes.extend(cols)
    
    # Filtrar dataset con solo columnas importantes
    if todas_importantes:
        df_filtrado = df[todas_importantes].copy()
        print(f"   ‚úÖ Dataset filtrado: {len(df_filtrado.columns)} columnas importantes")
    else:
        print(f"   ‚ö†Ô∏è No se encontraron columnas importantes, usando todas")
        df_filtrado = df.copy()
    
    # Identificar columnas de coordenadas
    columnas_coord = [col for col in df_filtrado.columns if any(palabra in col.upper() for palabra in ['COORD_ESTE', 'COORD_NORTE', 'ESTE', 'NORTE'])]
    
    if len(columnas_coord) >= 2:
        print(f"   üó∫Ô∏è Columnas de coordenadas: {columnas_coord}")
        
        # Convertir coordenadas a num√©rico
        for col in columnas_coord:
            df_filtrado[col] = pd.to_numeric(df_filtrado[col], errors='coerce')
        
        # Filtrar registros con coordenadas v√°lidas
        coords_antes = len(df_filtrado)
        
        # Eliminar registros con coordenadas nulas o (0,0)
        mask_valido = True
        for col in columnas_coord:
            mask_valido &= df_filtrado[col].notna()
            mask_valido &= (df_filtrado[col] != 0)
        
        # Filtrar por territorio peruano (aproximado)
        if 'COORD_ESTE' in str(columnas_coord[0]).upper() or 'ESTE' in str(columnas_coord[0]).upper():
            # Coordenadas UTM para Per√∫
            mask_valido &= (df_filtrado[columnas_coord[0]] >= 160000) & (df_filtrado[columnas_coord[0]] <= 1900000)
            mask_valido &= (df_filtrado[columnas_coord[1]] >= 8000000) & (df_filtrado[columnas_coord[1]] <= 10000000)
        
        df_filtrado = df_filtrado[mask_valido].copy()
        coords_despues = len(df_filtrado)
        
        print(f"   üåç Filtro de coordenadas:")
        print(f"     ‚Ä¢ Antes: {coords_antes:,} registros")
        print(f"     ‚Ä¢ Despu√©s: {coords_despues:,} registros")
        print(f"     ‚Ä¢ Eliminados: {coords_antes - coords_despues:,} ({(coords_antes - coords_despues)/coords_antes*100:.1f}%)")
        
        if coords_despues > 0:
            for col in columnas_coord:
                print(f"     ‚Ä¢ Rango {col}: {df_filtrado[col].min():.0f} - {df_filtrado[col].max():.0f}")
    
    else:
        print(f"   ‚ùå No se encontraron columnas de coordenadas v√°lidas")
        print(f"   üîç Columnas disponibles: {list(df_filtrado.columns)}")
        return None
    
    # Limpiar campos de texto
    for col in df_filtrado.columns:
        if df_filtrado[col].dtype == 'object':
            df_filtrado[col] = df_filtrado[col].astype(str).str.strip().str.upper()
    
    # Informaci√≥n final
    print(f"   ‚úÖ Procesamiento completado:")
    print(f"     ‚Ä¢ Registros finales: {len(df_filtrado):,}")
    print(f"     ‚Ä¢ Columnas finales: {len(df_filtrado.columns)}")
    print(f"     ‚Ä¢ Porcentaje conservado: {len(df_filtrado)/original_filas*100:.1f}%")
    
    return df_filtrado, columnas_importantes

# Procesar cada dataset
datasets_procesados = {}
for nombre, df in datasets_oefa.items():
    resultado = procesar_dataset_oefa(df, nombre)
    if resultado is not None:
        df_procesado, columnas_info = resultado
        datasets_procesados[nombre] = {
            'dataframe': df_procesado,
            'columnas_info': columnas_info
        }
```
