```python
# üìö IMPORTAR LIBRER√çAS Y CONFIGURACI√ìN INICIAL
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
import os
import struct
warnings.filterwarnings('ignore')

# Configurar pandas para mejor visualizaci√≥n
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_rows', 20)

print("üîß Librer√≠as cargadas correctamente")
print(f"üìÖ An√°lisis realizado: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# Verificar estructura de archivos
data_path = '../DATA/Padron_web_20251001'
print(f"\nüìÅ EXPLORANDO CARPETA: {data_path}")

try:
    archivos = os.listdir(data_path)
    print(f"‚úÖ Carpeta encontrada con {len(archivos)} archivos:")
    for i, archivo in enumerate(archivos, 1):
        size = os.path.getsize(os.path.join(data_path, archivo)) / 1024**2
        print(f"   {i}. {archivo} ({size:.1f} MB)")
except Exception as e:
    print(f"‚ùå Error accediendo a la carpeta: {e}")
```

```python
# üîß FUNCI√ìN PARA LEER ARCHIVOS DBF
def leer_dbf_header(filepath):
    """
    Lee el header de un archivo DBF para obtener informaci√≥n b√°sica
    """
    try:
        with open(filepath, 'rb') as f:
            # Leer header del DBF
            header = f.read(32)
            
            # Extraer informaci√≥n b√°sica
            version = header[0]
            last_update = (header[1], header[2], header[3])  # YY, MM, DD
            num_records = struct.unpack('<I', header[4:8])[0]
            header_length = struct.unpack('<H', header[8:10])[0]
            record_length = struct.unpack('<H', header[10:12])[0]
            
            # Calcular n√∫mero de campos
            num_campos = (header_length - 33) // 32
            
            return {
                'registros': num_records,
                'campos': num_campos,
                'tama√±o_registro': record_length,
                'tama√±o_header': header_length
            }
    except Exception as e:
        return {'error': str(e)}

# üìä AN√ÅLISIS DE CADA ARCHIVO DBF
print("üìä AN√ÅLISIS DETALLADO DE ARCHIVOS DBF")
print("=" * 55)

archivos_dbf = [
    ('Padron_web.dbf', 'Padr√≥n principal de instituciones educativas'),
    ('Padlocaladi_web.dbf', 'Datos locales adicionales'),
    ('Instituciones_apoyo.dbf', 'Instituciones de apoyo educativo')
]

info_archivos = {}

for archivo, descripcion in archivos_dbf:
    filepath = os.path.join(data_path, archivo)
    print(f"\nüîç ANALIZANDO: {archivo}")
    print(f"   üìù Descripci√≥n: {descripcion}")
    
    if os.path.exists(filepath):
        size_mb = os.path.getsize(filepath) / 1024**2
        print(f"   üì¶ Tama√±o: {size_mb:.1f} MB")
        
        # Leer informaci√≥n del header
        info = leer_dbf_header(filepath)
        info_archivos[archivo] = info
        
        if 'error' not in info:
            print(f"   üìä Registros: {info['registros']:,}")
            print(f"   üìã Campos: {info['campos']}")
            print(f"   üíæ Bytes por registro: {info['tama√±o_registro']}")
        else:
            print(f"   ‚ùå Error: {info['error']}")
    else:
        print(f"   ‚ùå Archivo no encontrado")

print(f"\nüìà RESUMEN DE ARCHIVOS:")
total_registros = sum([info.get('registros', 0) for info in info_archivos.values() if 'error' not in info])
print(f"   ‚Ä¢ Total de registros en todos los archivos: {total_registros:,}")
print(f"   ‚Ä¢ Archivo principal (Padron_web.dbf): {info_archivos.get('Padron_web.dbf', {}).get('registros', 0):,} registros")
print(f"   ‚Ä¢ Archivos exitosamente analizados: {len([i for i in info_archivos.values() if 'error' not in i])}/3")
```

```python
# üìö CARGAR DATOS EDUCATIVOS CON DBFREAD
from dbfread import DBF
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
import os
warnings.filterwarnings('ignore')

print("üîß Librer√≠a dbfread instalada y cargada")
print(f"üìÖ Cargando datos: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# Cargar el archivo principal de padr√≥n de instituciones educativas
data_path = '../DATA/Padron_web_20251001'
padron_file = os.path.join(data_path, 'Padron_web.dbf')

print(f"\nüìñ CARGANDO PADR√ìN PRINCIPAL DE INSTITUCIONES EDUCATIVAS...")
print(f"   üìÅ Archivo: {padron_file}")

try:
    # Leer DBF y convertir a DataFrame
    print(f"   üîÑ Leyendo archivo DBF...")
    table = DBF(padron_file, encoding='latin-1')
    
    print(f"   üîÑ Convirtiendo a DataFrame...")
    df_educacion = pd.DataFrame(iter(table))
    
    print(f"   ‚úÖ Datos cargados exitosamente!")
    print(f"   üìä Dimensiones: {df_educacion.shape[0]:,} filas √ó {df_educacion.shape[1]} columnas")
    print(f"   üíæ Memoria utilizada: {df_educacion.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
    
except Exception as e:
    print(f"   ‚ùå Error cargando datos: {e}")
    df_educacion = None
```

```python
# üîç EXPLORACI√ìN INICIAL DEL DATASET EDUCATIVO
print("üîç EXPLORACI√ìN INICIAL - PADR√ìN DE INSTITUCIONES EDUCATIVAS")
print("=" * 65)

print(f"üìä INFORMACI√ìN GENERAL:")
print(f"   ‚Ä¢ Total de instituciones educativas: {df_educacion.shape[0]:,}")
print(f"   ‚Ä¢ Total de campos/columnas: {df_educacion.shape[1]}")
print(f"   ‚Ä¢ Memoria utilizada: {df_educacion.memory_usage(deep=True).sum() / 1024**2:.1f} MB")

print(f"\nüìù COLUMNAS DISPONIBLES (primeras 30):")
for i, col in enumerate(df_educacion.columns[:30], 1):
    print(f"   {i:2d}. {col}")

if len(df_educacion.columns) > 30:
    print(f"   ... y {len(df_educacion.columns) - 30} columnas m√°s")

print(f"\nüìä TIPOS DE DATOS:")
tipos_count = df_educacion.dtypes.value_counts()
for tipo, count in tipos_count.items():
    print(f"   ‚Ä¢ {tipo}: {count} columnas")

print(f"\nüîç MUESTRA DE DATOS (primeras 3 filas, primeras 10 columnas):")
muestra = df_educacion.iloc[:3, :10]
print(muestra)
```

```python
# üéì AN√ÅLISIS POR NIVELES Y MODALIDADES EDUCATIVAS
print("üéì AN√ÅLISIS DETALLADO DE INSTITUCIONES EDUCATIVAS")
print("=" * 55)

print(f"üìä DISTRIBUCI√ìN POR NIVEL EDUCATIVO:")
if 'D_NIV_MOD' in df_educacion.columns:
    nivel_counts = df_educacion['D_NIV_MOD'].value_counts()
    print(f"   ‚Ä¢ Total de niveles educativos: {nivel_counts.count()}")
    for i, (nivel, count) in enumerate(nivel_counts.head(10).items(), 1):
        porcentaje = (count / len(df_educacion)) * 100
        print(f"   {i:2d}. {nivel}: {count:,} ({porcentaje:.1f}%)")

print(f"\nüè´ DISTRIBUCI√ìN POR MODALIDAD/FORMA:")
if 'D_FORMA' in df_educacion.columns:
    forma_counts = df_educacion['D_FORMA'].value_counts()
    for i, (forma, count) in enumerate(forma_counts.items(), 1):
        porcentaje = (count / len(df_educacion)) * 100
        print(f"   {i}. {forma}: {count:,} ({porcentaje:.1f}%)")

print(f"\nüèõÔ∏è DISTRIBUCI√ìN POR GESTI√ìN:")
if 'D_GESTION' in df_educacion.columns:
    gestion_counts = df_educacion['D_GESTION'].value_counts()
    for i, (gestion, count) in enumerate(gestion_counts.items(), 1):
        porcentaje = (count / len(df_educacion)) * 100
        print(f"   {i}. {gestion}: {count:,} ({porcentaje:.1f}%)")

print(f"\nüë• DISTRIBUCI√ìN POR TIPO DE SEXO:")
if 'D_TIPSSEXO' in df_educacion.columns:
    sexo_counts = df_educacion['D_TIPSSEXO'].value_counts()
    for i, (sexo, count) in enumerate(sexo_counts.items(), 1):
        porcentaje = (count / len(df_educacion)) * 100
        print(f"   {i}. {sexo}: {count:,} ({porcentaje:.1f}%)")

print(f"\nüåç DISTRIBUCI√ìN GEOGR√ÅFICA (Top 10 departamentos):")
if 'D_DPTO' in df_educacion.columns:
    dpto_counts = df_educacion['D_DPTO'].value_counts()
    print(f"   ‚Ä¢ Total de departamentos: {dpto_counts.count()}")
    for i, (dpto, count) in enumerate(dpto_counts.head(10).items(), 1):
        porcentaje = (count / len(df_educacion)) * 100
        print(f"   {i:2d}. {dpto}: {count:,} ({porcentaje:.1f}%)")
```

```python
# üó∫Ô∏è AN√ÅLISIS DE COORDENADAS Y UBICACI√ìN GEOGR√ÅFICA
print("üó∫Ô∏è AN√ÅLISIS DE UBICACI√ìN GEOGR√ÅFICA")
print("=" * 45)

# Buscar columnas que puedan contener coordenadas
coord_columns = [col for col in df_educacion.columns if any(keyword in col.lower() for keyword in ['coord', 'lat', 'lon', 'x_', 'y_', 'utm', 'este', 'norte'])]
print(f"üîç Columnas de coordenadas encontradas: {coord_columns}")

# Verificar otras columnas geogr√°ficas importantes
geo_columns = [col for col in df_educacion.columns if any(keyword in col.lower() for keyword in ['ubigeo', 'codgeo', 'local', 'centro', 'poblado'])]
print(f"üó∫Ô∏è Columnas geogr√°ficas importantes: {geo_columns[:10]}")

# Analizar √°rea urbana/rural
print(f"\nüèòÔ∏è DISTRIBUCI√ìN URBANA/RURAL:")
if 'DAREACENSO' in df_educacion.columns:
    area_counts = df_educacion['DAREACENSO'].value_counts()
    for i, (area, count) in enumerate(area_counts.items(), 1):
        porcentaje = (count / len(df_educacion)) * 100
        print(f"   {i}. {area}: {count:,} ({porcentaje:.1f}%)")

# Analizar centros poblados
print(f"\nüèòÔ∏è AN√ÅLISIS DE CENTROS POBLADOS:")
if 'CEN_POB' in df_educacion.columns:
    # Contar cu√°ntas instituciones tienen centro poblado registrado
    con_centro = df_educacion['CEN_POB'].notna().sum()
    sin_centro = len(df_educacion) - con_centro
    print(f"   ‚Ä¢ Con centro poblado registrado: {con_centro:,} ({(con_centro/len(df_educacion)*100):.1f}%)")
    print(f"   ‚Ä¢ Sin centro poblado: {sin_centro:,} ({(sin_centro/len(df_educacion)*100):.1f}%)")
    
    if con_centro > 0:
        print(f"\n   üèÜ Top 10 centros poblados con m√°s instituciones:")
        centros_top = df_educacion['CEN_POB'].value_counts().head(10)
        for i, (centro, count) in enumerate(centros_top.items(), 1):
            print(f"     {i:2d}. {centro}: {count:,} instituciones")

# An√°lisis de c√≥digos geogr√°ficos
print(f"\nüìç C√ìDIGOS GEOGR√ÅFICOS (UBIGEO):")
if 'CODGEO' in df_educacion.columns:
    codgeo_counts = df_educacion['CODGEO'].value_counts()
    print(f"   ‚Ä¢ Total de c√≥digos UBIGEO √∫nicos: {codgeo_counts.count():,}")
    print(f"   ‚Ä¢ Instituciones con UBIGEO: {df_educacion['CODGEO'].notna().sum():,}")

# Verificar datos faltantes en ubicaci√≥n
print(f"\nüí° COMPLETITUD DE DATOS GEOGR√ÅFICOS:")
geo_cols_check = ['D_DPTO', 'D_PROV', 'D_DIST', 'CODGEO', 'CEN_POB']
for col in geo_cols_check:
    if col in df_educacion.columns:
        missing = df_educacion[col].isnull().sum()
        complete = len(df_educacion) - missing
        porcentaje = (complete / len(df_educacion)) * 100
        print(f"   ‚Ä¢ {col}: {complete:,} completos ({porcentaje:.1f}%)")
        if missing > 0:
            print(f"     ‚îî‚îÄ Faltantes: {missing:,} ({(missing/len(df_educacion)*100):.1f}%)")
```

```python
# üìç AN√ÅLISIS DETALLADO DE COORDENADAS GEOGR√ÅFICAS
print("üìç AN√ÅLISIS DE COORDENADAS GEOGR√ÅFICAS")
print("=" * 45)

if 'NLAT_IE' in df_educacion.columns and 'NLONG_IE' in df_educacion.columns:
    # Analizar coordenadas disponibles
    lat_disponibles = df_educacion['NLAT_IE'].notna().sum()
    lon_disponibles = df_educacion['NLONG_IE'].notna().sum()
    
    print(f"üìä DISPONIBILIDAD DE COORDENADAS:")
    print(f"   ‚Ä¢ Latitud disponible: {lat_disponibles:,} ({(lat_disponibles/len(df_educacion)*100):.1f}%)")
    print(f"   ‚Ä¢ Longitud disponible: {lon_disponibles:,} ({(lon_disponibles/len(df_educacion)*100):.1f}%)")
    
    # Filtrar instituciones con coordenadas v√°lidas
    df_con_coords = df_educacion.dropna(subset=['NLAT_IE', 'NLONG_IE'])
    df_sin_coords = df_educacion[df_educacion[['NLAT_IE', 'NLONG_IE']].isnull().any(axis=1)]
    
    print(f"   ‚Ä¢ Instituciones CON coordenadas: {len(df_con_coords):,}")
    print(f"   ‚Ä¢ Instituciones SIN coordenadas: {len(df_sin_coords):,}")
    
    if len(df_con_coords) > 0:
        print(f"\nüåç RANGO GEOGR√ÅFICO DE INSTITUCIONES EDUCATIVAS:")
        print(f"   üìç Latitud: {df_con_coords['NLAT_IE'].min():.6f} a {df_con_coords['NLAT_IE'].max():.6f}")
        print(f"   üìç Longitud: {df_con_coords['NLONG_IE'].min():.6f} a {df_con_coords['NLONG_IE'].max():.6f}")
        
        print(f"\nüìä MUESTRA DE INSTITUCIONES CON COORDENADAS:")
        muestra_coords = df_con_coords[['CEN_EDU', 'D_NIV_MOD', 'D_DPTO', 'CEN_POB', 'NLAT_IE', 'NLONG_IE']].head(10)
        print(muestra_coords)
        
        # An√°lisis por departamento de instituciones con coordenadas
        print(f"\nüìç INSTITUCIONES CON COORDENADAS POR DEPARTAMENTO:")
        dept_coords = df_con_coords['D_DPTO'].value_counts()
        for i, (dept, count) in enumerate(dept_coords.head(10).items(), 1):
            total_dept = df_educacion[df_educacion['D_DPTO'] == dept].shape[0]
            porcentaje = (count / total_dept) * 100
            print(f"   {i:2d}. {dept}: {count:,} de {total_dept:,} ({porcentaje:.1f}%)")
        
        print(f"\nüíæ GUARDANDO INSTITUCIONES EDUCATIVAS CON COORDENADAS...")
        try:
            # Crear dataset de instituciones educativas con coordenadas
            instituciones_coords = df_con_coords[['CEN_EDU', 'D_NIV_MOD', 'D_GESTION', 'D_DPTO', 'D_PROV', 'D_DIST', 'CEN_POB', 'NLAT_IE', 'NLONG_IE']].copy()
            instituciones_coords.columns = ['Nombre_Institucion', 'Nivel_Modalidad', 'Gestion', 'Departamento', 'Provincia', 'Distrito', 'Centro_Poblado', 'Latitud', 'Longitud']
            
            instituciones_coords.to_csv('../DATA/instituciones_educativas_coordenadas.csv', index=False)
            print(f"   ‚úÖ Archivo guardado: ../DATA/instituciones_educativas_coordenadas.csv")
            print(f"   üìä {len(instituciones_coords):,} instituciones educativas con coordenadas")
        except Exception as e:
            print(f"   ‚ùå Error guardando: {e}")
            
else:
    print("‚ùå No se encontraron columnas de coordenadas NLAT_IE/NLONG_IE")
```

```python
# üéØ PROCESAMIENTO FINAL PARA HACKATON - DATASET EDUCACI√ìN
print("üéØ PROCESAMIENTO FINAL PARA HACKATON - INSTITUCIONES EDUCATIVAS")
print("=" * 70)

print("üìã SELECCIONANDO COLUMNAS ESENCIALES PARA LA HACKATON...")

# Filtrar solo instituciones con coordenadas (esenciales para el an√°lisis geoespacial)
df_educacion_final = df_educacion.dropna(subset=['NLAT_IE', 'NLONG_IE']).copy()

print(f"‚úÖ Instituciones educativas con coordenadas v√°lidas: {len(df_educacion_final):,}")

# Verificar rangos de coordenadas
print(f"\nüîç RANGOS ACTUALES DE COORDENADAS:")
print(f"   ‚Ä¢ Latitud (NLAT_IE): {df_educacion_final['NLAT_IE'].min():.6f} a {df_educacion_final['NLAT_IE'].max():.6f}")
print(f"   ‚Ä¢ Longitud (NLONG_IE): {df_educacion_final['NLONG_IE'].min():.6f} a {df_educacion_final['NLONG_IE'].max():.6f}")

# Seleccionar y renombrar columnas esenciales para la hackaton
columnas_esenciales = {
    'COD_MOD': 'codigo_modular',
    'CEN_EDU': 'nombre_institucion',
    'D_NIV_MOD': 'nivel_modalidad',
    'D_FORMA': 'forma_atencion',
    'D_GESTION': 'gestion',
    'D_DPTO': 'departamento',
    'D_PROV': 'provincia',
    'D_DIST': 'distrito',
    'CODGEO': 'ubigeo',
    'CEN_POB': 'centro_poblado',
    'DAREACENSO': 'area_censal',
    'NLAT_IE': 'latitud',
    'NLONG_IE': 'longitud',
    'D_TIPSSEXO': 'tipo_sexo',
    'DIR_CEN': 'direccion'
}

# Crear dataset final con solo las columnas necesarias
df_educacion_procesado = df_educacion_final[list(columnas_esenciales.keys())].copy()
df_educacion_procesado = df_educacion_procesado.rename(columns=columnas_esenciales)

print(f"\nüìä DATASET PROCESADO:")
print(f"   ‚Ä¢ Registros: {len(df_educacion_procesado):,}")
print(f"   ‚Ä¢ Columnas: {len(df_educacion_procesado.columns)}")
print(f"   ‚Ä¢ Memoria: {df_educacion_procesado.memory_usage(deep=True).sum() / 1024**2:.1f} MB")

print(f"\nüìù COLUMNAS FINALES:")
for i, col in enumerate(df_educacion_procesado.columns, 1):
    non_null = df_educacion_procesado[col].notna().sum()
    completitud = (non_null / len(df_educacion_procesado)) * 100
    print(f"   {i:2d}. {col}: {non_null:,} registros ({completitud:.1f}% completo)")

# Limpiar y estandarizar datos
print(f"\nüßπ LIMPIEZA Y ESTANDARIZACI√ìN:")

# Verificar y limpiar coordenadas (est√°n en formato decimal para Per√∫)
coords_validas_antes = len(df_educacion_procesado)

# Filtro para coordenadas v√°lidas en territorio peruano
df_educacion_procesado = df_educacion_procesado[
    (df_educacion_procesado['longitud'] >= -81.5) & 
    (df_educacion_procesado['longitud'] <= -68.0) &
    (df_educacion_procesado['latitud'] >= -18.5) & 
    (df_educacion_procesado['latitud'] <= 0.5) &
    (df_educacion_procesado['longitud'] != 0) &
    (df_educacion_procesado['latitud'] != 0)
]

coords_validas_despues = len(df_educacion_procesado)
print(f"   ‚úÖ Filtro de coordenadas v√°lidas (Per√∫): {coords_validas_despues:,} de {coords_validas_antes:,}")

# Limpiar campos de texto
df_educacion_procesado['nombre_institucion'] = df_educacion_procesado['nombre_institucion'].str.strip().str.upper()
df_educacion_procesado['departamento'] = df_educacion_procesado['departamento'].str.strip().str.upper()
df_educacion_procesado['provincia'] = df_educacion_procesado['provincia'].str.strip().str.upper()
df_educacion_procesado['distrito'] = df_educacion_procesado['distrito'].str.strip().str.upper()
df_educacion_procesado['centro_poblado'] = df_educacion_procesado['centro_poblado'].str.strip().str.upper()

# Estandarizar categor√≠as
df_educacion_procesado['nivel_modalidad'] = df_educacion_procesado['nivel_modalidad'].str.strip()
df_educacion_procesado['gestion'] = df_educacion_procesado['gestion'].str.strip()
df_educacion_procesado['forma_atencion'] = df_educacion_procesado['forma_atencion'].str.strip()

print(f"   ‚úÖ Campos de texto estandarizados (may√∫sculas, sin espacios extra)")
print(f"   ‚úÖ Categor√≠as normalizadas")

print(f"\nüìä ESTAD√çSTICAS FINALES:")
print(f"   ‚Ä¢ Departamentos √∫nicos: {df_educacion_procesado['departamento'].nunique()}")
print(f"   ‚Ä¢ Provincias √∫nicas: {df_educacion_procesado['provincia'].nunique()}")
print(f"   ‚Ä¢ Niveles educativos: {df_educacion_procesado['nivel_modalidad'].nunique()}")
print(f"   ‚Ä¢ Tipos de gesti√≥n: {df_educacion_procesado['gestion'].nunique()}")

print(f"\nüó∫Ô∏è RANGO GEOGR√ÅFICO FINAL:")
print(f"   ‚Ä¢ Longitud: {df_educacion_procesado['longitud'].min():.6f} a {df_educacion_procesado['longitud'].max():.6f}")
print(f"   ‚Ä¢ Latitud: {df_educacion_procesado['latitud'].min():.6f} a {df_educacion_procesado['latitud'].max():.6f}")

print(f"\nüìã MUESTRA DEL DATASET FINAL:")
print(df_educacion_procesado.head())
```
