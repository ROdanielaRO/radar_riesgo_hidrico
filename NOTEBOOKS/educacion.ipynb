# 🎓 Análisis de Instituciones Educativas - Padrón Web MINEDU
## Exploración del dataset de instituciones educativas del Perú

Este notebook contiene el análisis del **Padrón Web MINEDU** (fecha: 2025-10-01) con información completa de instituciones educativas del país.

### 📁 Estructura de archivos encontrados:
- `Padron_web.dbf` - Dataset principal de instituciones educativas
- `Padlocaladi_web.dbf` - Datos locales adicionales 
- `Instituciones_apoyo.dbf` - Instituciones de apoyo educativo
- `Especificacion de la tabla de datos padron web.xlsx` - Diccionario de datos principal
- `Especificacion de la tabla de datos locales adicionales.xlsx` - Diccionario de datos adicionales

```python
# 📚 IMPORTAR LIBRERÍAS Y CONFIGURACIÓN INICIAL
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
import os
import struct
warnings.filterwarnings('ignore')

# Configurar pandas para mejor visualización
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_rows', 20)

print("🔧 Librerías cargadas correctamente")
print(f"📅 Análisis realizado: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# Verificar estructura de archivos
data_path = '../DATA/Padron_web_20251001'
print(f"\n📁 EXPLORANDO CARPETA: {data_path}")

try:
    archivos = os.listdir(data_path)
    print(f"✅ Carpeta encontrada con {len(archivos)} archivos:")
    for i, archivo in enumerate(archivos, 1):
        size = os.path.getsize(os.path.join(data_path, archivo)) / 1024**2
        print(f"   {i}. {archivo} ({size:.1f} MB)")
except Exception as e:
    print(f"❌ Error accediendo a la carpeta: {e}")
```

```python
# 🔧 FUNCIÓN PARA LEER ARCHIVOS DBF
def leer_dbf_header(filepath):
    """
    Lee el header de un archivo DBF para obtener información básica
    """
    try:
        with open(filepath, 'rb') as f:
            # Leer header del DBF
            header = f.read(32)
            
            # Extraer información básica
            version = header[0]
            last_update = (header[1], header[2], header[3])  # YY, MM, DD
            num_records = struct.unpack('<I', header[4:8])[0]
            header_length = struct.unpack('<H', header[8:10])[0]
            record_length = struct.unpack('<H', header[10:12])[0]
            
            # Calcular número de campos
            num_campos = (header_length - 33) // 32
            
            return {
                'registros': num_records,
                'campos': num_campos,
                'tamaño_registro': record_length,
                'tamaño_header': header_length
            }
    except Exception as e:
        return {'error': str(e)}

# 📊 ANÁLISIS DE CADA ARCHIVO DBF
print("📊 ANÁLISIS DETALLADO DE ARCHIVOS DBF")
print("=" * 55)

archivos_dbf = [
    ('Padron_web.dbf', 'Padrón principal de instituciones educativas'),
    ('Padlocaladi_web.dbf', 'Datos locales adicionales'),
    ('Instituciones_apoyo.dbf', 'Instituciones de apoyo educativo')
]

info_archivos = {}

for archivo, descripcion in archivos_dbf:
    filepath = os.path.join(data_path, archivo)
    print(f"\n🔍 ANALIZANDO: {archivo}")
    print(f"   📝 Descripción: {descripcion}")
    
    if os.path.exists(filepath):
        size_mb = os.path.getsize(filepath) / 1024**2
        print(f"   📦 Tamaño: {size_mb:.1f} MB")
        
        # Leer información del header
        info = leer_dbf_header(filepath)
        info_archivos[archivo] = info
        
        if 'error' not in info:
            print(f"   📊 Registros: {info['registros']:,}")
            print(f"   📋 Campos: {info['campos']}")
            print(f"   💾 Bytes por registro: {info['tamaño_registro']}")
        else:
            print(f"   ❌ Error: {info['error']}")
    else:
        print(f"   ❌ Archivo no encontrado")

print(f"\n📈 RESUMEN DE ARCHIVOS:")
total_registros = sum([info.get('registros', 0) for info in info_archivos.values() if 'error' not in info])
print(f"   • Total de registros en todos los archivos: {total_registros:,}")
print(f"   • Archivo principal (Padron_web.dbf): {info_archivos.get('Padron_web.dbf', {}).get('registros', 0):,} registros")
print(f"   • Archivos exitosamente analizados: {len([i for i in info_archivos.values() if 'error' not in i])}/3")
```

```python
# 📚 CARGAR DATOS EDUCATIVOS CON DBFREAD
from dbfread import DBF
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
import os
warnings.filterwarnings('ignore')

print("🔧 Librería dbfread instalada y cargada")
print(f"📅 Cargando datos: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# Cargar el archivo principal de padrón de instituciones educativas
data_path = '../DATA/Padron_web_20251001'
padron_file = os.path.join(data_path, 'Padron_web.dbf')

print(f"\n📖 CARGANDO PADRÓN PRINCIPAL DE INSTITUCIONES EDUCATIVAS...")
print(f"   📁 Archivo: {padron_file}")

try:
    # Leer DBF y convertir a DataFrame
    print(f"   🔄 Leyendo archivo DBF...")
    table = DBF(padron_file, encoding='latin-1')
    
    print(f"   🔄 Convirtiendo a DataFrame...")
    df_educacion = pd.DataFrame(iter(table))
    
    print(f"   ✅ Datos cargados exitosamente!")
    print(f"   📊 Dimensiones: {df_educacion.shape[0]:,} filas × {df_educacion.shape[1]} columnas")
    print(f"   💾 Memoria utilizada: {df_educacion.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
    
except Exception as e:
    print(f"   ❌ Error cargando datos: {e}")
    df_educacion = None
```

```python
# 🗺️ ANÁLISIS DE COORDENADAS Y UBICACIÓN GEOGRÁFICA
print("🗺️ ANÁLISIS DE UBICACIÓN GEOGRÁFICA")
print("=" * 45)

# Buscar columnas que puedan contener coordenadas
coord_columns = [col for col in df_educacion.columns if any(keyword in col.lower() for keyword in ['coord', 'lat', 'lon', 'x_', 'y_', 'utm', 'este', 'norte'])]
print(f"🔍 Columnas de coordenadas encontradas: {coord_columns}")

# Verificar otras columnas geográficas importantes
geo_columns = [col for col in df_educacion.columns if any(keyword in col.lower() for keyword in ['ubigeo', 'codgeo', 'local', 'centro', 'poblado'])]
print(f"🗺️ Columnas geográficas importantes: {geo_columns[:10]}")

# Analizar área urbana/rural
print(f"\n🏘️ DISTRIBUCIÓN URBANA/RURAL:")
if 'DAREACENSO' in df_educacion.columns:
    area_counts = df_educacion['DAREACENSO'].value_counts()
    for i, (area, count) in enumerate(area_counts.items(), 1):
        porcentaje = (count / len(df_educacion)) * 100
        print(f"   {i}. {area}: {count:,} ({porcentaje:.1f}%)")

# Analizar centros poblados
print(f"\n🏘️ ANÁLISIS DE CENTROS POBLADOS:")
if 'CEN_POB' in df_educacion.columns:
    # Contar cuántas instituciones tienen centro poblado registrado
    con_centro = df_educacion['CEN_POB'].notna().sum()
    sin_centro = len(df_educacion) - con_centro
    print(f"   • Con centro poblado registrado: {con_centro:,} ({(con_centro/len(df_educacion)*100):.1f}%)")
    print(f"   • Sin centro poblado: {sin_centro:,} ({(sin_centro/len(df_educacion)*100):.1f}%)")
    
    if con_centro > 0:
        print(f"\n   🏆 Top 10 centros poblados con más instituciones:")
        centros_top = df_educacion['CEN_POB'].value_counts().head(10)
        for i, (centro, count) in enumerate(centros_top.items(), 1):
            print(f"     {i:2d}. {centro}: {count:,} instituciones")

# Análisis de códigos geográficos
print(f"\n📍 CÓDIGOS GEOGRÁFICOS (UBIGEO):")
if 'CODGEO' in df_educacion.columns:
    codgeo_counts = df_educacion['CODGEO'].value_counts()
    print(f"   • Total de códigos UBIGEO únicos: {codgeo_counts.count():,}")
    print(f"   • Instituciones con UBIGEO: {df_educacion['CODGEO'].notna().sum():,}")

# Verificar datos faltantes en ubicación
print(f"\n💡 COMPLETITUD DE DATOS GEOGRÁFICOS:")
geo_cols_check = ['D_DPTO', 'D_PROV', 'D_DIST', 'CODGEO', 'CEN_POB']
for col in geo_cols_check:
    if col in df_educacion.columns:
        missing = df_educacion[col].isnull().sum()
        complete = len(df_educacion) - missing
        porcentaje = (complete / len(df_educacion)) * 100
        print(f"   • {col}: {complete:,} completos ({porcentaje:.1f}%)")
        if missing > 0:
            print(f"     └─ Faltantes: {missing:,} ({(missing/len(df_educacion)*100):.1f}%)")
```

```python
# 🎯 PROCESAMIENTO FINAL PARA HACKATON - DATASET EDUCACIÓN
print("🎯 PROCESAMIENTO FINAL PARA HACKATON - INSTITUCIONES EDUCATIVAS")
print("=" * 70)

print("📋 SELECCIONANDO COLUMNAS ESENCIALES PARA LA HACKATON...")

# Filtrar solo instituciones con coordenadas (esenciales para el análisis geoespacial)
df_educacion_final = df_educacion.dropna(subset=['NLAT_IE', 'NLONG_IE']).copy()

print(f"✅ Instituciones educativas con coordenadas válidas: {len(df_educacion_final):,}")

# Verificar rangos de coordenadas
print(f"\n🔍 RANGOS ACTUALES DE COORDENADAS:")
print(f"   • Latitud (NLAT_IE): {df_educacion_final['NLAT_IE'].min():.6f} a {df_educacion_final['NLAT_IE'].max():.6f}")
print(f"   • Longitud (NLONG_IE): {df_educacion_final['NLONG_IE'].min():.6f} a {df_educacion_final['NLONG_IE'].max():.6f}")

# Seleccionar y renombrar columnas esenciales para la hackaton
columnas_esenciales = {
    'COD_MOD': 'codigo_modular',
    'CEN_EDU': 'nombre_institucion',
    'D_NIV_MOD': 'nivel_modalidad',
    'D_FORMA': 'forma_atencion',
    'D_GESTION': 'gestion',
    'D_DPTO': 'departamento',
    'D_PROV': 'provincia',
    'D_DIST': 'distrito',
    'CODGEO': 'ubigeo',
    'CEN_POB': 'centro_poblado',
    'DAREACENSO': 'area_censal',
    'NLAT_IE': 'latitud',
    'NLONG_IE': 'longitud',
    'D_TIPSSEXO': 'tipo_sexo',
    'DIR_CEN': 'direccion'
}

# Crear dataset final con solo las columnas necesarias
df_educacion_procesado = df_educacion_final[list(columnas_esenciales.keys())].copy()
df_educacion_procesado = df_educacion_procesado.rename(columns=columnas_esenciales)

print(f"\n📊 DATASET PROCESADO:")
print(f"   • Registros: {len(df_educacion_procesado):,}")
print(f"   • Columnas: {len(df_educacion_procesado.columns)}")
print(f"   • Memoria: {df_educacion_procesado.memory_usage(deep=True).sum() / 1024**2:.1f} MB")

print(f"\n📝 COLUMNAS FINALES:")
for i, col in enumerate(df_educacion_procesado.columns, 1):
    non_null = df_educacion_procesado[col].notna().sum()
    completitud = (non_null / len(df_educacion_procesado)) * 100
    print(f"   {i:2d}. {col}: {non_null:,} registros ({completitud:.1f}% completo)")

# Limpiar y estandarizar datos
print(f"\n🧹 LIMPIEZA Y ESTANDARIZACIÓN:")

# Verificar y limpiar coordenadas (están en formato decimal para Perú)
coords_validas_antes = len(df_educacion_procesado)

# Filtro para coordenadas válidas en territorio peruano
df_educacion_procesado = df_educacion_procesado[
    (df_educacion_procesado['longitud'] >= -81.5) & 
    (df_educacion_procesado['longitud'] <= -68.0) &
    (df_educacion_procesado['latitud'] >= -18.5) & 
    (df_educacion_procesado['latitud'] <= 0.5) &
    (df_educacion_procesado['longitud'] != 0) &
    (df_educacion_procesado['latitud'] != 0)
]

coords_validas_despues = len(df_educacion_procesado)
print(f"   ✅ Filtro de coordenadas válidas (Perú): {coords_validas_despues:,} de {coords_validas_antes:,}")

# Limpiar campos de texto
df_educacion_procesado['nombre_institucion'] = df_educacion_procesado['nombre_institucion'].str.strip().str.upper()
df_educacion_procesado['departamento'] = df_educacion_procesado['departamento'].str.strip().str.upper()
df_educacion_procesado['provincia'] = df_educacion_procesado['provincia'].str.strip().str.upper()
df_educacion_procesado['distrito'] = df_educacion_procesado['distrito'].str.strip().str.upper()
df_educacion_procesado['centro_poblado'] = df_educacion_procesado['centro_poblado'].str.strip().str.upper()

# Estandarizar categorías
df_educacion_procesado['nivel_modalidad'] = df_educacion_procesado['nivel_modalidad'].str.strip()
df_educacion_procesado['gestion'] = df_educacion_procesado['gestion'].str.strip()
df_educacion_procesado['forma_atencion'] = df_educacion_procesado['forma_atencion'].str.strip()

print(f"   ✅ Campos de texto estandarizados (mayúsculas, sin espacios extra)")
print(f"   ✅ Categorías normalizadas")

print(f"\n📊 ESTADÍSTICAS FINALES:")
print(f"   • Departamentos únicos: {df_educacion_procesado['departamento'].nunique()}")
print(f"   • Provincias únicas: {df_educacion_procesado['provincia'].nunique()}")
print(f"   • Niveles educativos: {df_educacion_procesado['nivel_modalidad'].nunique()}")
print(f"   • Tipos de gestión: {df_educacion_procesado['gestion'].nunique()}")

print(f"\n🗺️ RANGO GEOGRÁFICO FINAL:")
print(f"   • Longitud: {df_educacion_procesado['longitud'].min():.6f} a {df_educacion_procesado['longitud'].max():.6f}")
print(f"   • Latitud: {df_educacion_procesado['latitud'].min():.6f} a {df_educacion_procesado['latitud'].max():.6f}")

print(f"\n📋 MUESTRA DEL DATASET FINAL:")
print(df_educacion_procesado.head())
```
